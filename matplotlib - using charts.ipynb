```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


final_result = pd.read_excel('python-excel.xlsx')
final_result
```

```python
plt.style.use('seaborn')
final_result.plot(kind="line", marker="o", title="MTH overall", x="Student Name", y="MTH 102", color="red", figsize=(15, 5))
plt.show()
```

```python
plt.style.use('seaborn')
final_result_sorted = final_result.sort_values('BIO 101')
final_result_sorted.plot(kind="bar", title="BIO 101 OVERALL", x="Student Name", y="BIO 101", color="orange", figsize=(15,10))
plt.show()
```

```python
final_result.sort_values("BIO 101").plot(kind="barh", title="BIO 101 OVERALL", x="Student Name", y="BIO 101", color="red", figsize=(18,10))
plt.show()
```

```python
#final_result.plot(kind="pie", title="BIO 101 OVERALL", y="BIO 101",figsize=(18,10))
students = final_result["Student Name"]
course = final_result["BIO 101"]
```

```python
plt.pie(course, labels=students, autopct="1%.1f%%")
plt.show()
```

```python
color = ["red", "lightblue", "tomato", "chocolate"]
final_result.sort_values("BIO 101").plot(kind="bar", title="BIO 101 OVERALL", x="Student Name", figsize=(20,10), color=color)
plt.show()
```
